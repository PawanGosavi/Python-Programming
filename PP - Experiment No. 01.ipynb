### **P1 - C**

### **Aim** : Write a program to generate the Fibonacci series.

```python
# input
nterms = int(input("\nEnter The Term till You want to print Fibonacci Sequence : "))

# first two terms
n1 = 0
n2 = 1
count = 0
fslist = []

# check if the number of terms is valid
if nterms <= 0:
   print("\nInvalid Number, Please enter a Positive Integer")

# if there is only one term, return n1
elif nterms == 1:
   print("\nFibonacci Sequence upto", nterms, ": ")
   print(n1)

# generate fibonacci sequence
else:
    while count < nterms:
       # print("")
       # print(n1)
       fslist.append(n1)
       # print(fslist)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
    # Fibonacci Sequence
    print("\nFibonacci Sequence upto", nterms, ": ", fslist)
```

### **P1 - D**

### **Aim** : Write a function that reverses the user defined value.

```python
# input
number = int(input("\nEnter The Number You want to Reverse : "))

# check if the number of terms is valid
if number <= 0:
   print("\nInvalid Number, Please enter a Positive Integer")

else:
   reversed_number = 0

   while number != 0:
       reminder = number % 10
       reversed_number = reversed_number * 10 + reminder
       number //= 10

   print("\nReversed Number : " + str(reversed_number))
```

### **P1 - E**

### **Aim** : Write a function to check the input value is Armstrong and also write the function for Palindrome.

```python
def Armstrong(number):
   # check if the number of terms is valid
   if number <= 0:
      print("\nInvalid Number, Please enter a Positive Integer")

   else:
      # Changed num variable to string, 
      # and calculated the length (number of digits)
      order = len(str(number))

      # initialize sum
      number_sum = 0

      # find the sum of the cube of each digit
      temp = number
      while temp > 0:
         digit = temp % 10
         number_sum += digit ** order
         temp //= 10

      print("")
      # display the result
      if number == number_sum:
         print(str(number) + " is an Armstrong Number")
      else:
         print(str(number) + " is not an Armstrong Number")

# for i in range (0, 201):
#   Armstrong(i)

# input
number = int(input("\nEnter The Number You want to Check as Armstrong Number : "))
if number != "":
   Armstrong(number)
```

```python
def Palindrome(sequence):
   text = str(sequence).lower()
   print("")
   # display the result
   if text == text[::-1]:
      print(str(sequence) + " is a Palindrome.")
   else:
      print(str(sequence) + " is not a Palindrome.")

# input
text = input("\nEnter The Text / Number You want to Check as Palindrome : ")
if text != "":
   Palindrome(text)
```

### **P1 - F**

### **Aim** : Write a recursive function to print the factorial for a given number.

```python
def recur_factorial(n):
   if n == 1:
       return n
   else:
       return n*recur_factorial(n-1)

# input
number = int(input("\nEnter The Number You want to Check for Factorial : "))

# check if the number is negative
if number < 0:
   print("\nSorry, factorial does not exist for negative numbers")
elif number == 0:
   print("\nThe factorial of 0 is 1")
else:
   print("\nThe factorial of", number, "is", recur_factorial(number))
```
