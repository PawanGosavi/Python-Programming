### **P6 - B**

### **Aim** : Write a Python program to append text to a file and display the text.

```python
path = "/content/drive/MyDrive/demo/readme.txt"

print("---------------- Before ----------------")
f = open(path,'r')
print("File       : "+ f.name)
print("Permission : " + f.mode + "\n")
result = f.read()
print(result)
f.close()

f = open(path,'a+')
f.write('I live Near Kalamboli.\n')
f.close()

print("---------------- After ----------------")
f = open(path,'r')
print("File       : "+ f.name)
print("Permission : " + f.mode + "\n")
result = f.read()
print(result)
f.close()
```

### **P6 - C**

### **Aim** : Write a Python program to read last n lines of a file.

```python
path = "/content/drive/MyDrive/demo/readme.txt"

f = open(path,'r')
print("File       : "+ f.name)
print("Permission : " + f.mode + "\n")
lines = f.readlines()
count = int(input("How Many Lines You want to Print from File? : "))
print("")
result = lines[-count:]
for i in result:
    print(i)
f.close()
```
