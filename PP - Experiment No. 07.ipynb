### **P7 - B**

### **Aim** : Implement the concept of inheritance using python.

```python
class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputsides(self):
        self.sides = [float(input("\nEnter side "+str(i+1)+" : ")) for i in range(self.n)]

    def displaysides(self):
        for i in range(self.n):
            print("\nSide",i+1," : ",self.sides[i])

class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self, 3)
        
    def findarea(self):
        a, b, c = self.sides
        # calculate the semi-perimeter
        s = (a + b + c) / 2
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print("\nThe Area of the Triangle is : %0.2f" %area)

# Creating an Object
print("\n------ Finding The Area of the Triangle! ------")
t = Triangle()

# Providing input for sides
t.inputsides()

# Display sides
t.displaysides()

# Print results
t.findarea()
```

### **P7 - C**

### **Aim** : Create a class called Numbers, which has a single class attribute called MULTIPLIER, and a constructor which takes the parameters x and y (these should all be numbers).

### a. Write a method called add which returns the sum of the attributes x and y.
### b. Write a class method called multiply, which takes a single number parameter a and returns the product of a and MULTIPLIER.
### c. Write a static method called subtract, which takes two number parameters, b and c, and returns b - c.
### d. Write a method called value which returns a tuple containing the values of x and y. Make this method into a property, and write a setter and a deleter for manipulating the values of x and y.

```python
class Numbers:
    MULTIPLIER = 3.5

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        return self.x + self.y

    @classmethod
    def multiply(cls, a):
        return cls.MULTIPLIER * a

    @staticmethod
    def subtract(b, c):
        return b - c

    @property
    def value(self):
        return (self.x, self.y)

    @value.setter
    def value(self, xy_tuple):
        self.x, self.y = xy_tuple

    @value.deleter
    def value(self):
        del self.x
        del self.y

# Taking Inputs
print("\nFor Math Operations, Enter 2 Random Numbers")
num1 = int(input("\nEnter Number 1 : "))
num2 = int(input("\nEnter Number 2 : "))
           
# Creating Object
n = Numbers(num1, num2)

# Doing Addition using Return
print("\nAddition of", num1, "and", num2, "is", n.add())

# Doing Multiplication using Class Method
print("\nMultiplication of", n.MULTIPLIER, "and", num2, "is", n.multiply(num2))

# Doing Subtration using Static Method
print("\nSubtraction of", num1, "and", num2, "is", n.subtract(num1, num2))

# Converting Value to Tumple using Setter
print("\nThe", num1, "and", num2, "now becomes tuple : ", n.value)
```
