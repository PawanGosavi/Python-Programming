### **P5 - A**

### **Aim** : Write a Python script to sort (ascending and descending) a dictionary by value.

```python
# import files
 
import operator

n = int(input("\nHow Many Number of Elements You want in Dictionary? : \t"))
d = {}

for i in range(n):
    print("\nDictionary Member %d :" %i)
    keys = input("Key :\t")
    values = int(input("Value :\t"))
    d[keys] = values

print("\nDictionary : ", d)

ascending_d = dict(sorted(d.items(), key = operator.itemgetter(1)))

print("\nDictionary [Sorted : Ascending] : ", ascending_d)

descending_d = dict(sorted(d.items(), key = operator.itemgetter(1), reverse = True))

print("\nDictionary [Sorted : Descending] : ", descending_d)
```

### **P5 - B**

### **Aim** : Write a Python script to concatenate following dictionaries to create a new one.

Sample Dictionary :

dic1={1:10, 2:20},

dic2={3:30, 4:40},

dic3={5:50, 6:60},

Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

```python
dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

dic4 = {}

print("\nDictionary 1 : ", dic1)
print("\nDictionary 2 : ", dic2)
print("\nDictionary 3 : ", dic3)

def update_dict(inputd, targetd):
    print("\nConcatinating Dictionary", inputd, "with Dictionary", targetd)
    targetd.update(inputd)
    print("\nUpdated Dictionary : ", targetd)

for i in (dic1, dic2, dic3):
    update_dict(i, dic4)

print("\n\nDictionary : ", dic4)
```

### **P5 - C**

### **Aim** : Write a Python program to sum all the items in a dictionary.

```python
n = int(input("\nHow Many Number of Elements You want in Dictionary? : \t"))
d = {}

for i in range(n):
    print("\nDictionary Member %d :" %i)
    keys = input("Key :\t")
    values = int(input("Value :\t"))
    d[keys] = values

print("\nDictionary : ", d)

print("\nThe Sum of All Items in Dictionary is : ", sum(d.values()))
```
