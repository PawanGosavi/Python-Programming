### **P4 - A**

### **Aim** : Write a program that takes two lists and returns True if they have at least one common member.

```python
# import files
 
import sys
import string as str
 
def commonmemberstr(list1, list2):
    list1 = list1.split(", ")
    list2 = list2.split(", ")
    cm = []
    for i in list1:
        if i in list2:
            cm.append(i)
    if len(cm) != 0:
        print("\nBoth The List have at least 1 Common Member? : True")
        print("\nTotal", len(cm), "Common Elements Found, Those are :", cm)
 
    else:
        print("\nBoth The List have at least 1 Common Member? : False")
 
print("\nEnter List elements to Check for Common Member.")
list1 = input("\nEnter Elements for List 1 : ")
list2 = input("\nEnter Elements for List 2 : ")
 
commonmemberstr(list1, list2)
```

### **Aim** : Write a Python program to clone or copy a list.

```python
l = input("\nEnter Elements for List to Clone or Copy : ")
l = l.split(", ")

def listclone(l):
    print("\nGiven List : ", l)
    try:
        copy_l = l.copy()
        print("\nCopied List : ", copy_l)
    except:
        clone_l = l
        print("\nCloned List : ", clone_l)
    else:
        pass
    finally:
        if len(l) == len(copy_l):
            print("\nList Copied Successfully.")
        elif len(l) == len(clone_l):
            print("\nList Cloned Successfully.")
        else:
            print("\nList Operation was not Successful.")

listclone(l)
```
