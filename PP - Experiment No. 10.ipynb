### **P10 - A**

### **Aim** : Design a simple database application that stores the records and retrieve the same.

```python
# Database Connection
db = sqlite3.connect("syit.db")
print("\nDatabase Successfully Created.")
cursor = db.cursor()
```

```python
# Table Deletion
cursor.execute("drop table if exists students")
```

```python
# Table Creation
try:
    cursor.execute("""create table students(id int auto_increment primary key,
    name varchar(25), email varchar(25), age varchar(5), mobile varchar(10))""")
    print("\nTable Successfully Created.")
except:
    print("\nAlready table existed !!")
```

```python
# Show Tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
# Insert Queries
sql = "insert into students(name, email, age, mobile) values(?, ?, ?, ?)"

val = [
	('Pawan', 'pawan@gmail.com', '23', '8796135214'),
	('Rohan', 'rohan@gmail.com', '22', '8796135211'),
	('Sohan', 'sohan@gmail.com', '21', '8796135212'),
	('Mohan', 'mohan@gmail.com', '24', '8796135213')
]

cursor.executemany(sql, val)
db.commit()

print(cursor.rowcount, "Records were inserted.")
```

```python
# Print Tables
cursor.execute("select * from students")

result = cursor.fetchall()

# print(result)

for x in result:
    print(x)
```

```python
# Database Connection
db = sqlite3.connect("syit.db")
# print("\nDatabase Successfully Created.")
cursor = db.cursor()
```

```python
# Show Tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
# Print Tables
cursor.execute("select * from students")

result = cursor.fetchall()

# print(result)

for x in result:
    print(x)
```

```python
# Add Records

sql = "insert into students(name, email, age, mobile) values(?, ?, ?, ?)"

val = [
	('Karan', 'karan@gmail.com', '20', '8796135216'),
	('Kiran', 'kiran@gmail.com', '17', '8796135215'),
]

cursor.executemany(sql, val)
db.commit()

print(cursor.rowcount, "Records were inserted.")
```

```python
# Print Tables
cursor.execute("select * from students")

result = cursor.fetchall()

# print(result)

for x in result:
    print(x)
```

```python
# Delete Records

cursor.execute("delete from students where age < 18")
db.commit()

print(cursor.rowcount, "Records were Deleted.")
```

```python
# Print Tables
cursor.execute("select * from students")

result = cursor.fetchall()

# print(result)

for x in result:
    print(x)
```
