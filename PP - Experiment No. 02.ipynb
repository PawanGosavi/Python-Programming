### **P2 - A**

### **Aim** : Write a function that takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
def stringvowelcheck(string):
    vowels = ["a", "e", "i", "o", "u"]
    s = string.lower()
    uc = []
    vl = []
    for i in s:
        if i not in uc:
            if i != " ":
                uc.append(i)
    print("\nCharacters : ", uc)
    for c in uc:
        if c in vowels:
            print("\nCharacter : " + c + ", Is a Vowel? : True")
            vl.append(c)
        else:
            print("\nCharacter : " + c + ", Is a Vowel? : False")
    print("\nTotal", str(len(vl)), "Vowels Found, Those are :", vl)

# Input
c = input("\nEnter Character to Check if Vowel or Not : ")

stringvowelcheck(c)
```
